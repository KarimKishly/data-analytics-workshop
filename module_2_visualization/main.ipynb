```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot grades vs year
year = [2020, 2021, 2022, 2023, 2024]
grades = [18, 17.5, 17.75, 17, 18.5]

plt.plot(year, grades)
# Set labels
plt.xlabel('Year')
plt.ylabel('Grade')
# Set the ticks on the x-axis
plt.xticks(year)
plt.title('Grade vs Year')
plt.show()
```

```python
data = pd.read_csv('diabetes.csv')
data
```

```python
# Plot glucose vs age in our dataset using matplotlib function
X = data['age'].values
y = data['glucose'].values

plt.scatter(X, y)
plt.show()
```

```python
# Plot glucose vs age in our dataset using pandas function
data[['age', 'glucose']].plot(title='Grade vs Year',kind='scatter', x='age', y='glucose')
```

```python
# Plot histogram of age distribution in the dataset
data['age'].hist()
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
